# Part 5: Brownian Motion

## Normal Distribution

回顾：一元正态分布的概率密度函数为：$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$

将一元正态分布扩展到二维，我们得到二维正态分布。令$(X_1, X_2) \sim N_2(\mu, \Sigma)$，其中$\mu = \begin{pmatrix} \mu_1 \\ \mu_2 \end{pmatrix}$是均值向量，$\Sigma = \begin{pmatrix} \sigma_1^2 & \rho\sigma_1\sigma_2 \\ \rho\sigma_1\sigma_2 & \sigma_2^2 \end{pmatrix}$是协方差矩阵。二维正态分布的概率密度函数为：$$f(x_1, x_2) = \frac{1}{2\pi\sqrt{|\Sigma|}} e^{-\frac{1}{2} (X - \mu)^T \Sigma^{-1} (X - \mu)}$$

- 其中$\rho_i^2 = \operatorname{Var}X_i$, $\rho_{\sigma_1 \sigma_2} = \operatorname{Cov}(X_1, X_2)$.

再次扩展到多维，令$X = (X_1, X_2, \ldots, X_n) \sim N_n(\mu, \Sigma)$，其中$\mu$是$n$维均值向量，$\Sigma$是$n \times n$协方差矩阵。多维正态分布的概率密度函数为：$$f(x) = \frac{1}{(2\pi)^{n/2} |\Sigma|^{1/2}} e^{-\frac{1}{2} (X - \mu)^T \Sigma^{-1} (X - \mu)}$$

### Marginal Distribution

多维正态分布可以分为多个边缘分布。例如，若$X = (X_1, X_2, \ldots, X_n) \sim N_n(\mu, \Sigma)$，并将$X$分为两部分$X_1 = (X_1, \ldots, X_k)$和$X_2 = (X_{k+1}, \ldots, X_n)$，则$X_1$和$X_2$分别服从边缘正态分布：$$X_1 \sim N_k(\mu_1, \Sigma_{11})$$ $$X_2 \sim N_{n-k}(\mu_2, \Sigma_{22})$$

- 另外，$\Sigma_{12}(i, j) = \text{cov}(X_1(i), X_2(j))$，其中 $1 \leq i \leq p_1$ 和 $1 \leq j \leq p_2$.

### Conditional Distribution

多维正态分布的条件分布也服从正态分布。继续上述例子，给定$X_2 = x_2$，则$X_1$的条件分布为：$$X_1 | X_2 = x_2 \sim N_k(\mu_{1|2}, \Sigma_{1|2})$$

- 其中，$$\mu_{1|2} = \mu_1 + \Sigma_{12} \Sigma_{22}^{-1} (x_2 - \mu_2)$$ $$\Sigma_{1|2} = \Sigma_{11} - \Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21}$$
- 推导时可使用条件概率定义：$f(x_1 | x_2) = \frac{f(x_1, x_2)}{f(x_2)}$.


## Brownian Motion

布朗运动可以被定义为简单随机游走的连续极限。考虑$\mathbb{Z}$上的随机游走$$X_n = Y_1 + Y_2 + \cdots + Y_n,$$ 其中$Y_i$是独立同分布的随机变量，取值为$+1$或$-1$，概率各为$1/2$。

通过适当缩放时间和空间，我们可以得到连续极限（即布朗运动）。由于时间间隔为$1/n$，可以得到：$$B_t^{(n)} = \frac{1}{\sqrt{n}} X_{\lfloor nt \rfloor}$$
- 可以看出，当时间间隔$n$加速一个因子时，空间间隔需要缩小一个因子$\sqrt{n}$以保持运动的性质。经过每一个时间间隔$1/n$，位置变化为$\pm 1/\sqrt{n}$。

### Limiting Process

当$n \to \infty$时，过程$B_t^{(n)}$收敛到布朗运动$B_t$。布朗运动具有以下性质：
1. 期望为零：$E[B_t] = 0$。
2. 方差与时间成正比：$\operatorname{Var}(B_t) = t$。
3. 正态分布：$B_t \sim N(0, t)$。可由CLT推导得到。
4. 增量独立：对于任意$0 \leq s < t$，增量$B_t - B_s$与过去的运动无关。
5. 增量服从正态分布：$B_t - B_s \sim N(0, t - s)$。 因此，$B_t - B_s$与$B_{t-s}$在分布上相同。

标准布朗运动（Standard Brownian Motion）是指初始位置为零的布朗运动，即$B_0 = 0$。

方差参数$\sigma^2$的布朗运动可以表示为：$B_t^{(\sigma)} = \sigma B_t$. 它具备一个标准布朗运动的所有性质，但其分布为$N(0, \sigma^2 t)$。

### Basic Properties

1. **Markov性质**: 对于 $t > s$，给定 $B_s = x$，则 $B_t \sim \mathcal{N}(x, t-s)$，其概率密度函数为 $$f_t(y|B_s=x) = \frac{1}{\sqrt{2\pi(t-s)}} \exp\left(-\frac{1}{2}\frac{(y-x)^2}{t-s}\right).$$ 
2. **尺度变换性质**: 对于所有 $c > 0$，$\{cB_{t/c^2}\}_{t\geq 0}$ 是一个标准布朗运动。 
3. **对称性**: $\{-B_t\}_{t\geq 0}$ 是一个标准布朗运动。 
4. **时间反转性质**: 对于固定的 $T > 0$，定义 $Y_t = B_T - B_{T-t}$，则 $\{Y_t\}_{0 \leq t \leq T}$ 是一个标准布朗运动。

回顾：高斯过程是指任意有限维度的边缘分布均为多维正态分布的随机过程，其性质包括：

1. 任意线性组合仍为高斯过程。
2. 完全由均值函数和协方差函数决定。

布朗运动是一个高斯过程，因为对于任意时间点集合$t_1, t_2, \ldots, t_n$，随机变量$(B_{t_1}, B_{t_2}, \ldots, B_{t_n})$服从多维正态分布。

## Hitting Times & Maximum Variable

### Hitting Times

布朗运动的首次到达时间（Hitting Time）定义为：$$T_a = \inf\{t \geq 0 : B_t = a\}$$
- 其中$\inf$表示下确界（即最小的达到时间），$a$为目标位置。

**反射原理（Reflection Principle）**：令$B_t$为一个标准布朗运动，则有：$$ B_t^* = \begin{cases} B_t, & \text{if } t \le T_a \\ 2a - B_t, & \text{if } t > T_a \end{cases} $$ 也是一个标准布朗运动。

### Maximum Variable

定义布朗运动在时间$t$内的最大值为：$M_t = \max_{0 \leq s \leq t} B_s$.

联合分布：对于$a > 0$和$b < a$，有：$$\mathbb{P}(M_t \geq a, B_t \leq b) = \mathbb{P}(B_t \geq 2a - b) = 1 - \Phi\left(\frac{2a - b}{\sqrt{t}}\right)$$
- 其中$\Phi$为标准正态分布的累积分布函数。
- 证明：利用反射原理，事件$\{M_t \geq a, B_t \leq b\}$等价于$\{B_t^* \geq a, B_t \leq b\}$，而$B_t^*$与$B_t$在分布上相同。

边际分布：$$\mathbb{P}(M_t \geq a) = 2 \mathbb{P}(B_t \geq a) = 2\left(1 - \Phi\left(\frac{a}{\sqrt{t}}\right)\right)$$

首次到达时间$T_a$与${a^2}/{B_1^2}$具有相同的分布，其分布为：$$\mathbb{P}(T_a \leq t) = \mathbb{P}(M_t \geq a) = 2\left(1 - \Phi\left(\frac{|a|}{\sqrt{t}}\right)\right)$$

例：令$A, B > 0$，求$\mathbb{P}(T_A < T_{-B})$。

解：布朗运动为随机游走的极限，因此可将其转化为赌徒破产问题，即：考虑一个以$A$为赢利目标，$-B$为破产点的对称随机游走。  
回顾：设赌徒的初始资金为$i$，其在破产前达到目标$N$的概率为$p_i = \frac{i}{N}$。  
因此，布朗运动中，初始位置为$0$，目标为$A$，破产点为$-B$，则有：$$\mathbb{P}(T_A < T_{-B}) = \frac{B}{A + B}$$

## Variations of Brownian Motion

### BM with Drift

布朗运动的漂移版本定义为：$$X_t = \mu t + \sigma B_t$$
- 其中$\mu$为漂移率，$\sigma$为波动率，$B_t$为标准布朗运动。

其具有如下性质：
1. 期望：$E[X_t] = \mu t$。
2. 方差：$\operatorname{Var}(X_t) = \sigma^2 t$。
3. 分布：$X_t \sim N(\mu t, \sigma^2 t)$。

### Geometric Brownian Motion

几何布朗运动定义为：$$Y_t = Y_0 e^{(\mu - \frac{1}{2}\sigma^2)t + \sigma B_t}$$
- 其中$Y_0$为初始值，$\mu$为漂移率，$\sigma$为波动率，$B_t$为标准布朗运动。

其具有如下性质：
1. 期望：$E[Y_t] = Y_0 e^{\mu t}$。
2. 方差：$\operatorname{Var}(Y_t) = Y_0^2 e^{2\mu t}(e^{\sigma^2 t} - 1)$。
3. 分布：$\ln(Y_t) \sim N\left(\ln(Y_0) + (\mu - \frac{1}{2}\sigma^2)t, \sigma^2 t\right)$。
4. **条件期望**：$E[Y_t | Y_s] = Y_s e^{\mu (t - s)}$，对于$t > s$。

几何布朗运动可以作为股票价格的模型。其中，$\mu$表示股票的预期收益率，$\sigma$表示股票价格的波动性。

### Planar Brownian Motion

平面布朗运动定义为二维向量过程：$$\mathbf{B}_t = (B_t^{(1)}, B_t^{(2)})$$
- 其中$B_t^{(1)}$和$B_t^{(2)}$为独立的标准布朗运动。

平面布朗运动从原点开始，具有独立、平稳的高斯增量。

## Pricing Stock Options

### Black-Scholes Model

统计套利定价模型中，股票价格通常被建模为几何布朗运动。假设股票价格$S_t$遵循以下随机微分方程：$$dS_t = \mu S_t dt + \sigma S_t dB_t$$
- 其中$\mu$为股票的预期收益率，$\sigma$为股票的波动率，$B_t$为标准布朗运动。

该等式的解是一个几何布朗运动：$$S_t = S_0 e^{(\mu - \frac{1}{2}\sigma^2)t + \sigma B_t}$$

考虑一个股票期权（这里仅考虑看涨），则在时间$T$时，期权的价格为：$$\begin{cases} S_T - K, & \text{if } S_T > K \\ 0, & \text{if } S_T \leq K \end{cases} = (S_T - K)^+$$

进行期权定价时，我们希望购买期权的预期收益率与直接购买股票的预期收益率相同。即：$$E[e^{-rT}(Y_T - K)^+] = c.$$
- 其中$r = \mu$为无风险利率，$c$为期权的现值。

下面，求解期权的现值$c$。

由于$S_T$服从几何布朗运动，因此$\ln(S_T)$服从正态分布：$$\ln(S_T) \sim N\left(\ln(S_0) + (\mu - \frac{1}{2}\sigma^2)T, \sigma^2 T\right)$$
定义变量：$$d_1 = \frac{\ln(S_0/K) + (\mu + \frac{1}{2}\sigma^2)T}{\sigma \sqrt{T}}, \quad d_2 = d_1 - \sigma \sqrt{T}$$
则期权现值$c$可表示为：$$c = S_0 \Phi(d_1) - K e^{-rT} \Phi(d_2)$$
- 其中$\Phi$为标准正态分布的累积分布函数。
这个公式即为著名的Black-Scholes期权定价公式。

### Estimation of Volatility

估计股票价格的波动率$\sigma$是期权定价中的关键步骤。以下给出一个常用的方法：  

令${Y_t}$在时间$[0, T]$内具有漂移系数$\mu$和波动率$\sigma$。  

1. 将时间$[0, T]$划分为$n$个等间隔的时间点：$0 = t_0 < t_1 < \cdots < t_N = T$，其中$t_i = i\Delta t$，$\Delta t = T/N = h$。
2. 取$W_i = Y_{ih} - Y_{(i-1)h}$，则$W_i \sim N(\mu h, \sigma^2 h)$，且$W_i$相互独立。
3. 令$ S^2 = \frac{1}{N-1}\sum_{i=1}^N (W_i - \overline{W})^2, \quad \overline{W} = \frac{1}{N}\sum_{i=1}^N W_i $.
4. 可以看出，$\frac{(N-1)S^2}{\sigma^2 h} \sim \chi^2_{N-1}$，因此$E[S^2] = \sigma^2 h$。
5. 最后，波动率的估计量为：$$\hat{\sigma} = \frac{S}{\sqrt{h}} = \sqrt{\frac{1}{(N-1)h} \sum_{i=1}^N (W_i - \overline{W})^2}$$
